## 3. Data Loading and Initial Cleaning

This section loads the dataset and performs initial cleaning steps to ensure data quality before further analysis and model training. It includes handling missing values, standardizing columns, and removing irrelevant features.

```python
import pandas as pd
import numpy as np

# Load the dataset (using the path that appears to work based on previous execution context)
df = pd.read_csv('//content//Employee_Attrition - Raw_DataSet.csv')

print(f"Initial Total Employees: {df.shape[0]}")
print(f"Initial Total Features: {df.shape[1]}")

# Drop 'Employee_ID' if it exists, as it's typically not a predictive feature
if 'Employee_ID' in df.columns:
    df.drop(columns=['Employee_ID'], inplace=True)
    print("Dropped 'Employee_ID' column.")

# Drop duplicates
df.drop_duplicates(inplace=True)
print(f"Shape after dropping duplicates: {df.shape}")

# Handle 'Age' column
df['Age'] = df['Age'].abs() # Ensure age is positive
mean_age = round(df['Age'].mean())
df['Age'] = df['Age'].replace(0, mean_age) # Replace 0 with mean
df['Age'] = df['Age'].fillna(mean_age) # Fill NaN with mean
df['Age'] = df['Age'].round().astype(int) # Round and convert to int
print(f"Processed Age column. Mean age used: {mean_age}, Min age: {df['Age'].min()}")

# Handle 'Monthly_Income' column
df['Monthly_Income'] = pd.to_numeric(df['Monthly_Income'], errors='coerce')
df['Monthly_Income'] = df['Monthly_Income'].fillna(df['Monthly_Income'].median())
print("Processed 'Monthly_Income' column.")

# Handle other numerical columns ('Years_at_Company', 'Work_Life_Balance', etc.)
nums_to_fix = ['Years_at_Company', 'Work_Life_Balance', 'Job_Satisfaction', 'Job_Level', 'Distance_From_Home']
for col in nums_to_fix:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
print(f"Processed numerical columns: {', '.join(nums_to_fix)}.")

# Standardize binary columns like 'Attrition', 'Gender', 'Overtime', 'Marital_Status'
def standardize_binary(val):
    if pd.isna(val):
        return np.nan
    val = str(val).strip().lower()
    if val in ['yes', 'y', '1', 'true']:
        return 'Yes'
    elif val in ['no', 'n', '0', 'false']:
        return 'No'
    return np.nan

df['Attrition'] = df['Attrition'].apply(standardize_binary)

# Fill missing categorical values with mode for specified columns
cols_to_fix = ['Gender', 'Department', 'Job_Role', 'Marital_Status', 'Overtime'] # Attrition is handled by standardize_binary
for col in cols_to_fix:
    df[col] = df[col].fillna(df[col].mode()[0])
print(f"Processed categorical columns: {', '.join(cols_to_fix)}.")


print(f"\nFinal DataFrame shape: {df.shape}")
print("Missing values after cleaning:")
display(df.isnull().sum()[df.isnull().sum() > 0])
print("Unique Attrition values after standardization:")
print(df['Attrition'].unique())

print(f"\nTotal Employees: {df.shape[0]}")
print(f"Total Features: {df.shape[1]}")
print("\nInitial Attrition distribution:")
print(df['Attrition'].value_counts())
```

## 4. Key Visualizations after Data Cleaning

Now that the data is loaded and cleaned, let's re-run some key visualizations to understand the distributions and relationships.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['Attrition'] == 'Yes']['Years_at_Company'], label='Left Company', fill=True, color='red')
sns.kdeplot(data=df[df['Attrition'] == 'No']['Years_at_Company'], label='Stayed', fill=True, color='blue')
plt.title('Distribution of Tenure: Stayed vs. Left')
plt.xlabel('Years at Company')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 10))

# Only calculating correlation for numeric columns
# Ensure 'Attrition' is converted to numeric for correlation calculation
numeric_df_for_corr = df.copy()
numeric_df_for_corr['Attrition_Numeric'] = numeric_df_for_corr['Attrition'].map({'Yes': 1, 'No': 0})
numeric_df_for_corr = numeric_df_for_corr.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_df_for_corr.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
# Selecting a few key features to keep the pair plot readable
# Make sure 'Attrition' is in a suitable format for hue
features = ['Age', 'Monthly_Income', 'Years_at_Company', 'Average_Hours_Worked_Per_Week', 'Attrition']

sns.pairplot(df[features], hue='Attrition', corner=True, palette='husl')
plt.suptitle('Pair Plot of Key Features by Attrition', y=1.02) # Add a main title
plt.show()
```
